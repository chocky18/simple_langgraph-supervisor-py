```python
# Skincare Data
skincare_data = [
    {
        "concern": "acne",
        "treatment": "Use salicylic acid or benzoyl peroxide-based cleansers. Maintain a balanced diet and hydration.",
        "skincare_routine": "Cleanse twice daily, use a lightweight moisturizer, and apply SPF during the day."
    },
    {
        "concern": "dry skin",
        "treatment": "Use a hydrating cleanser and a thick moisturizer. Avoid hot showers.",
        "skincare_routine": "Moisturize immediately after washing your face. Use a humidifier in dry weather."
    }
]

# Nutrition Data
nutrition_data = [
    {
        "concern": "acne",
        "nutrition_advice": "Increase intake of vitamin A-rich foods like carrots and sweet potatoes. Stay hydrated and reduce sugar consumption."
    },
    {
        "concern": "dry skin",
        "nutrition_advice": "Consume omega-3 fatty acids (fish, flaxseeds) and drink plenty of water."
    }
]

# Save data as JSON files
with open("skincare.json", "w") as f:
    json.dump(skincare_data, f)
with open("nutrition.json", "w") as f:
    json.dump(nutrition_data, f)
```

```python
# Import necessary libraries
import faiss
import json
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from sentence_transformers import SentenceTransformer

# Load embedding model
embedding_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Load skincare data from JSON
with open("/content/skincare.json", "r") as f:
    skincare_data = json.load(f)

# Load nutrition data from JSON
with open("/content/nutrition.json", "r") as f:
    nutrition_data = json.load(f)

# Prepare text for embeddings
skincare_texts = [f"{entry['concern']} {entry['treatment']} {entry['skincare_routine']}" for entry in skincare_data]
nutrition_texts = [f"{entry['concern']} {entry['nutrition_advice']}" for entry in nutrition_data]

# Create FAISS vector stores
skincare_db = FAISS.from_texts(skincare_texts, embeddings)
nutrition_db = FAISS.from_texts(nutrition_texts, embeddings)

# Save FAISS indexes
skincare_db.save_local("skincare_faiss")
nutrition_db.save_local("nutrition_faiss")

print("✅ FAISS indexes for skincare and nutrition created successfully!")
```

```python
import faiss
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langgraph.prebuilt import create_react_agent
from langchain.tools import Tool

# ✅ Load FAISS Embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# ✅ Load FAISS Vector Stores
skincare_db = FAISS.load_local("skincare_faiss", embeddings, allow_dangerous_deserialization=True)
nutrition_db = FAISS.load_local("nutrition_faiss", embeddings, allow_dangerous_deserialization=True)

# ✅ Function to Query Skincare Database
def get_skincare_advice(query: str) -> str:
    """Fetches skincare advice from FAISS based on user concerns."""
    results = skincare_db.similarity_search(query, k=1)
    return results[0].page_content if results else "No skincare advice found."

def get_nutrition_advice(query: str) -> dict:
    """Fetches nutrition recommendations from FAISS for skin health."""
    results = nutrition_db.similarity_search(query, k=1)

    if results:
        return {"nutrition_advice": results[0].page_content}
    else:
        return {"nutrition_advice": "No nutrition advice found."}
```

```python
# ✅ Define Tools
skincare_tool = Tool(
    name="SkincareAdvisor",
    func=get_skincare_advice,
    description="Fetches skincare advice for concerns like acne, dryness, and pigmentation."
)

nutrition_tool = Tool(
    name="NutritionAdvisor",
    func=get_nutrition_advice,
    description="Fetches nutritional advice for skin health based on user concerns."
)

# ✅ Load LLaMA-3 Model (Replace with Groq API if needed)
from langchain_groq import ChatGroq

llm = ChatGroq(
    model_name="llama3-8b-8192",
    temperature=0.3,
    max_tokens=512
)

# ✅ Skincare Agent
skincare_agent = create_react_agent(
    model=llm,
    tools=[skincare_tool],
    name="skincare_expert",
    prompt="You are a dermatologist. Provide expert skincare advice using FAISS-based knowledge."
)

# ✅ Nutrition Agent
nutrition_agent = create_react_agent(
    model=llm,
    tools=[nutrition_tool],
    name="nutrition_expert",
    prompt="You are a nutritionist. Suggest personalized diet plans for skin health using FAISS-based retrieval."
)
```

```python
from langgraph_supervisor import create_supervisor

# ✅ Supervisor Agent
supervisor_agent = create_supervisor(
    [skincare_agent, nutrition_agent],
    model=llm,
    prompt=(
        "You are a supervisor managing skincare and nutrition experts. "
        "Route skin-related questions to skincare_agent and diet-related questions to nutrition_agent."
    )
)
```

```python
# ✅ Compile the Supervisor Workflow
app = supervisor_agent.compile()

# ✅ Run a Test Query
user_query = {"messages": [{"role": "user", "content": "i have acne suggest me skin care and nutrition"}]}
result = app.invoke(user_query)
print(result)
# ✅ Extract the Last AIMessage Response (Final Answer)
final_response = ""
for message in result["messages"]:
    if message.name == "supervisor":  # The last supervisor response contains the final summary
        final_response = message.content

# ✅ Print the Final Response
print("📝 Final Advice from Supervisor Agent:\n")
print(final_response)
```
